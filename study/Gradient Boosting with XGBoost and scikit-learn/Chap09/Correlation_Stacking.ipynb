#### 특성 공학 - 타임스탬프 데이터

```python
df['date'] = pd.to_datetime(df['time_stamp'])
df.head()

# 1970-01-01 00:25:44.952607890
# 우버는 1970년대에 없다 -> 오류
```

```python
pd.to_datetime(df['time_stamp'], unit='ms')
```

```python
df['date'] = pd.to_datetime(df['time_stamp']*(10**6))
df.head()

# *(10**6) 해결
```

```python
import datetime as dt
df['month'] = df['date'].dt.month
df['hour'] = df['date'].dt.hour
df['dayofweek'] = df['date'].dt.dayofweek


# dayofweek 으로 일주일 중 주말인지 파악 -> 주말에 관한 열 뽑아내기 -> 특성공학 
```

```python
def weekend(row):
    if row['dayofweek'] in [5,6]:
        return 1
    else:
        return 0

df['weekend'] = df.apply(weekend, axis=1)
```

```python
def rush_hour(row):
    if (row['hour'] in [6,7,8,9,15,16,17,18]) & (row['weekend'] == 0):
        return 1
    else:
        return 0

df['rush_hour'] = df.apply(rush_hour, axis=1)


# 시간으로 rush_hour : 출퇴근 시간 파악 가능
```

## 상관관계가 낮은 앙상블 만들기

### 다양한 모델

```python
from sklearn.datasets import load_breast_cancer

# 단순히 XGBoost나 랜덤 포레스트 같은 모델이 아니라
# 순수한 의미로 앙상블 (여러 모델을 합한 것)
```

```python
X, y = load_breast_cancer(return_X_y=True)
```

```python
kfold = StratifiedKFold(n_splits=5)
```

```python
from sklearn.model_selection import cross_val_score

def classification_model(model):
    # 5폴드 교차 검증을 수행합니다.
    scores = cross_val_score(model, X, y, cv=kfold)

    # 평균 점수를 반환합니다.
    return scores.mean()
```

```python
classification_model(XGBClassifier())

# 기본 XGB 분류기가 가장 성능이 좋다.
# 여러 모델 비교하는 과정 우리모델에서도 돌려보면 좋겠다. 
# 기본 학습기 바꿔가면서
```

```python
classification_model(XGBClassifier(booster='gblinear'))
```

```python
classification_model(XGBClassifier(booster='dart', one_drop=True))
```

```python
classification_model(RandomForestClassifier(random_state=2))
```

```python
classification_model(LogisticRegression(max_iter=10000))
```

```python
classification_model(XGBClassifier(n_estimators=500, max_depth=2, 
                                   learning_rate=0.1))
```



### VotingClassifier

```python
# xgb와의 상관관계가 비교적 적다고 파악된 로지스틱, 포레스트를 앙상블시키는 과정

# 1 - 빈 리스트를 만든다 
estimators = []

# 2 - 첫번째 모델 (로지스틱) 생성하고 더함
logistic_model = LogisticRegression(max_iter=10000)
estimators.append(('logistic', logistic_model))

# 3 - 두번째 모델 (XGB) 생성하고 리스트에 어펜드
xgb_model = XGBClassifier(max_depth=2, n_estimators=500, learning_rate=0.1)
estimators.append(('xgb', xgb_model))

# 4 - 세번째 모델 (랜덤 포레스트) 생성 및 어펜드
rf_model = RandomForestClassifier(random_state=2)
estimators.append(('rf', rf_model))

# 5 - VotingClassifier 을 통해 앙상블 모델 제작 
ensemble = VotingClassifier(estimators)
scores = cross_val_score(ensemble, X, y, cv=kfold)
print(scores.mean())

# 교차검증 수치 상승
```

```python
# 통합 코드

estimators = []
logistic_model = LogisticRegression(max_iter=10000)
estimators.append(('logistic', logistic_model))
xgb_model = XGBClassifier(max_depth=2, n_estimators=500, learning_rate=0.1)
estimators.append(('xgb', xgb_model))
rf_model = RandomForestClassifier(random_state=2)
estimators.append(('rf', rf_model))
ensemble = VotingClassifier(estimators)
scores = cross_val_score(ensemble, X, y, cv=kfold)
print(scores.mean())

# VotingClassifier
# 우리가 써야 하는건 Voting Regressor
```

## 스태킹

### StackingClassifier

```python
# 빈 리스트 생성  (앙상블 모델 제작 과정과 유사)
base_models = []

# 세 모델을 어펜드 ('이름', 모델) 튜플 형식으로 추가
base_models.append(('lr', LogisticRegression()))
base_models.append(('xgb', XGBClassifier()))
base_models.append(('rf', RandomForestClassifier(random_state=2)))

# 메타 모델을 정의합니다. (상위 모델로 간단한 모델을 활용하는 것이 좋다. )
meta_model = LogisticRegression()

# 스태킹 앙상블을 만듭니다.
clf = StackingClassifier(estimators=base_models, final_estimator=meta_model)
scores = cross_val_score(clf, X, y, cv=kfold)
print(scores.mean())

# 난 StackingRegressor을 활용하면 됨
```

```python
base_models = []
base_models.append(('lr', LogisticRegression()))
base_models.append(('xgb', XGBClassifier()))
base_models.append(('rf', RandomForestClassifier(random_state=2)))
# 메타 모델을 정의합니다.
meta_model = LogisticRegression()
# 스태킹 앙상블을 만듭니다.
clf = StackingClassifier(estimators=base_models, final_estimator=meta_model)
scores = cross_val_score(clf, X, y, cv=kfold)
print(scores.mean())
```
